```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
import os
```

```python
folder_path = '/Users/alina/Desktop/AFM/simulation_data' # /Users/alina/Desktop/AFM/simulation_data
file_str = 'eSentiment.csv'

eSentiment = pd.read_csv(os.path.join(folder_path, file_str))
```

```python
eSentiment_data = eSentiment.iloc[[7,9]+list(range(21,772))]
eSentiment_data
```

```python
initial_cols = eSentiment.iloc[20, :].to_list()
```

```python
eSentiment_data.columns = initial_cols
```

```python
returns = eSentiment_data.loc[:,[f for f in initial_cols if f == 'returns' or f == 'time']]
```

```python
returns = returns.reset_index(drop=True)
```

```python
time = range(1,752)
returns_02_025 = returns.iloc[2:,1].astype(float)
returns_02_05 = returns.iloc[2:,2].astype(float)
returns_02_075 = returns.iloc[2:,3].astype(float)
```

```python

# Plot returns, max news sensitivity, and max base propensity over time
plt.figure(figsize=(12, 6))

# Plot returns
plt.plot(time, returns_02_025, label='Returns BP=.25', color='black', linestyle='solid', linewidth=.2)

# Plot max news sensitivity
plt.plot(time, returns_02_05, label='Returns BP=0.50', color='blue', linestyle='dashdot', linewidth=.2)

# Plot max base propensity
plt.plot(time, returns_02_075, label='Returns BP=0.75', color='red', linestyle='dashed', linewidth=.2)

# Remove specific borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Comparison of Returns, Max News Sensitivity, and Max Base Propensity Over Time')
plt.legend()

# Save the plot at 300 DPI
#plt.savefig('/Users/alina/Desktop/AFM/figure/Returns_02.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```

```python
def graphing_varying_returns(time, return1, return2, return3, tosave=False, file_path: str=None, level: float=None):

    # Plot returns, max news sensitivity, and max base propensity over time
    plt.figure(figsize=(12, 6))

    # Plot returns
    plt.plot(time, return1, label='Returns BP=.25', color='black', linestyle='solid', linewidth=.2)

    # Plot max news sensitivity
    plt.plot(time, return2, label='Returns BP=0.50', color='blue', linestyle='dashdot', linewidth=.2)

    # Plot max base propensity
    plt.plot(time, return3, label='Returns BP=0.75', color='red', linestyle='dashed', linewidth=.2)

    # Remove specific borders
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # Add labels and title
    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.title(f'Comparison of Returns, Max News Sensitivity={level}, and Max Base Propensity Over Time')
    plt.legend()

    if tosave:
        # Save the plot at 300 DPI
        plt.savefig(file_path, dpi=300, bbox_inches='tight')
    else:
        pass

    # Display plot
    plt.show()
```
