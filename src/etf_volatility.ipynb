```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
import os
```

```python
folder_path = '/Users/alina/Desktop/AFM/simulation_data' # /Users/alina/Desktop/AFM/simulation_data
file_str = 'eSentiment.csv'

eSentiment = pd.read_csv(os.path.join(folder_path, file_str))
```

```python
eSentiment
```

```python
eSentiment_data = eSentiment.iloc[[7,9]+list(range(21,772))]
eSentiment_data
```

```python
initial_cols = eSentiment.iloc[20, :].to_list()
```

```python
eSentiment_data.columns = initial_cols
```

```python
volatility = eSentiment_data.loc[:,[f for f in initial_cols if f == 'volatility-indicator' or f == 'time']]
```

```python
volatility = volatility.reset_index(drop=True)
volatility
```

```python
volatility.iloc[2:,1].astype(float)
```

```python
time = range(1,752)
vol_02_025 = volatility.iloc[2:,1].astype(float)
vol_02_05 = volatility.iloc[2:,2].astype(float)
vol_02_075 = volatility.iloc[2:,3].astype(float)
```

```python

# Plot returns for varying max news sensitivity (mns) and max base propensity (bp) over time
plt.figure(figsize=(12, 6))

# returns at mns = 0.2 bs = 0.25
plt.plot(time, vol_02_025, label='BP=.25', color='black', linestyle='solid', linewidth=.2, marker='o', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='black', markersize=3)

# returns at mns = .2 bs = .5
plt.plot(time, vol_02_05, label='BP=0.50', color='blue', linestyle='dashdot', linewidth=.2, marker='s', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='blue', markersize=3)

# returns at mns = .2 bp = .75
plt.plot(time, vol_02_075, label='BP=0.75', color='red', linestyle='dashed', linewidth=.2, marker='<', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='red', markersize=3)

# Adjust the border line (spine) width
for spine in plt.gca().spines.values():
    spine.set_linewidth(0.3)  # Set spine (border) line width

# Remove specific borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# labels and title
plt.xlabel('Time', fontsize=8, labelpad=1) # reduce padding
plt.ylabel('Volatility', fontsize=8, labelpad=1)
plt.title('Comparison of Volatility, Max News Sensitivity, and Max Base Propensity Over Time')
plt.legend()

# Customize tick label font size
plt.tick_params(axis='both', which='major', labelsize=8) # reduce label size

# Save the plot at 300 DPI and COMMENT OUT when not saving
plt.savefig('/Users/alina/Desktop/AFM/figure/Volatility_02.png', dpi=300, bbox_inches='tight')

# Display plot
plt.show()
```

```python
def graphing_varying_volatility(time, list1, list2, list3, tosave=False, file_path: str=None, level: float=None):

    # Plot volatility at varying max news sensitivity (mns), and max base propensity (bp) over time
    plt.figure(figsize=(12, 6))

    # volatility 1 at mns =  bs = 
    plt.plot(time, list1, label='BP=.25', color='black', linestyle='solid', linewidth=.2,
              marker='o', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='black', markersize=3)

    # volatility 2 at mns = condition1 bp = condition2
    plt.plot(time, list2, label='BP=0.50', color='blue', linestyle='dashdot', linewidth=.2,
             marker='s', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='black', markersize=3)

    # Plot max base propensity
    plt.plot(time, list3, label='BP=0.75', color='red', linestyle='dashed', linewidth=.2,
             marker='<', markerfacecolor='none', markeredgewidth=.2, markeredgecolor='red', markersize=3)

    for spine in plt.gca().spines.values():
        spine.set_linewidth(.3)  # Set spine (border) line width

    # Remove specific borders
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)

    # Add labels and title
    plt.xlabel('Time')
    plt.ylabel('Volatility')
    plt.title(f'Comparison of Volatility, Max News Sensitivity={level}, and Max Base Propensity Over Time')
    plt.legend()

    # Customize tick label font size
    plt.tick_params(axis='both', which='major', labelsize=8) # reduce label size

    if tosave:
        # Save the plot at 300 DPI
        plt.savefig(file_path, dpi=300, bbox_inches='tight')
    else:
        pass

    # Display plot
    plt.show()
```
